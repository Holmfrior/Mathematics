##### Comparing and Plotting Pearson and Cosine Similarity

```python
import numpy as np
import matplotlib.pyplot as plt

# base
x = np.array([0, 1, 2, 3])

# offset
offsets = np.arange(-50, 51)

# initialize result
pearson_r = np.zeros(len(offsets))
cosine_sim = np.zeros(len(offsets))

# Process
for i, offset in enumerate(offsets):
    y = x + offset

    # Compute Pearson Correlation
    # Specific value at 0, 1
    pearson_r[i] = np.corrcoef(x, y)[0, 1]

    # Compute Cosine Similarity
    num = np.dot(x, y)
    den = np.linalg.norm(x) * np.linalg.norm(y)

    #  edge case: if den is 0, cosine similarity is undefined (set to 0)
    cosine_sim[i] = num / den if den != 0 else 0

# Plot
plt.figure(figsize=(10, 6))
plt.plot(offsets, pearson_r, 'ko-', label='Pearson', linewidth=1, markersize=5)
plt.plot(offsets, cosine_sim, 'gs-', label='Cosine sim.', color='gray', alpha=0.7)

plt.xlabel('Mean offset')
plt.ylabel('r or c')
plt.legend()
plt.grid(True, alpha=0.3)
plt.title('Correlation vs. Cosine Similarity with Varying Offset')
plt.show()
```

##### Time series filtering and feature detection.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define the Data ---
kernel = np.array([-1, 1])
signal = np.zeros(30)
signal[10:20] = 1.5

# --- 2. Perform the Calculation ---
feature_response = []
for i in range(len(signal) - 1):
    segment = signal[i : i+2]
    dot_prod = np.dot(kernel, segment)
    feature_response.append(dot_prod)

# --- 2.5 FIX: Adjust Alignment ---
# OLD WAY (Left Shifted):
# response_padded = np.append(feature_response, 0)

# NEW WAY (Aligned):
# We insert a 0 at the START. This shifts everything right by 1 index.
# This represents that at time=0, we haven't seen an edge yet.
response_padded = np.insert(feature_response, 0, 0)

# --- 3. Plotting ---
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Graph A: Kernel
axes[0].plot([14, 15], kernel, 'ks-', linewidth=1, markersize=5)
axes[0].set_title('A)\nKernel')
axes[0].set_xlim(0, 30)
axes[0].set_ylim(-1.1, 1.6)
axes[0].grid(True, alpha=0.3)

# Graph B: Time Series Signal
axes[1].plot(signal, 'ks-', linewidth=1, markersize=4)
axes[1].set_title('B)\nTime Series signal')
axes[1].set_xlim(0, 30)
axes[1].set_ylim(-1.1, 1.6)
axes[1].grid(True, alpha=0.3)

# Graph C: The Result
axes[2].plot(signal, 'ks-', label='Signal', linewidth=1, markersize=4)
# Note: The gray spike will now sit exactly at x=10 and x=20
axes[2].plot(response_padded, 'o-', label='Edge detection', color='gray', alpha=0.7, markersize=5)

axes[2].set_title('C)')
axes[2].legend(loc='lower left', fontsize='small', framealpha=1, edgecolor='gray')
axes[2].set_xlim(0, 30)
axes[2].set_ylim(-1.1, 1.6)
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

##### K-centroids

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Generate Mock Data (Matches the "Blobs" in your image) ---
np.random.seed(42) # For reproducible results
n_per_group = 50
# Create 3 distinct groups of data
group1 = np.random.randn(n_per_group, 2) + [2, 2]
group2 = np.random.randn(n_per_group, 2) + [-2, -2]
group3 = np.random.randn(n_per_group, 2) + [2, -2]
data = np.vstack((group1, group2, group3))

# --- Step 1: Initialize k Centroids ---
k = 3
# Randomly select k data samples to be the starting centroids
ridx = np.random.choice(range(len(data)), k, replace=False)
centroids = data[ridx, :]

# Setup for the 2x2 Plot (to match Figure 4-3)
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes = axes.flatten()

# --- Helper Function to Draw "Spokes" (Lines to Centroid) ---
def plot_kmeans_step(ax, data, centroids, groupidx, iteration_title):
    # 1. Plot the "Spokes" (Lines connecting points to their centroid)
    # We loop through unique groups to plot lines efficiently
    if groupidx is not None:
        for i in range(len(data)):
            # Get the centroid this point belongs to
            assignment = groupidx[i]
            center = centroids[assignment]
            point = data[i]
            # Plot line from point to center
            ax.plot([point[0], center[0]], [point[1], center[1]],
                    color='gray', alpha=0.3, linewidth=0.5, zorder=1)

    # 2. Plot the Data Points (Empty circles like the book)
    ax.scatter(data[:, 0], data[:, 1], facecolors='none', edgecolors='k', s=30, zorder=2)

    # 3. Plot the Centroids (Black dots)
    ax.scatter(centroids[:, 0], centroids[:, 1], c='black', s=100, zorder=3)

    ax.set_title(iteration_title, fontsize=14)
    ax.set_xticks([])
    ax.set_yticks([])

# --- Plot Iteration 0 (Initialization) ---
# We calculate initial distances just for the plot, but don't move centroids yet
dists_init = np.zeros((data.shape[0], k))
for ci in range(k):
    dists_init[:, ci] = np.sum((data - centroids[ci, :])**2, axis=1)
groupidx_init = np.argmin(dists_init, axis=1)
plot_kmeans_step(axes[0], data, centroids, groupidx_init, "Iteration 0")


# --- Step 5: The Loop (Iterate 3 times) ---
# The text specifies repeating Steps 2-4
for i in range(3):

    # --- Step 2: Compute Distances (Vectorized) ---
    dists = np.zeros((data.shape[0], k))
    for ci in range(k):
        # Subtract centroid from data, square, and sum columns
        dists[:, ci] = np.sum((data - centroids[ci, :])**2, axis=1)

    # --- Step 3: Assign Groups ---
    groupidx = np.argmin(dists, axis=1)

    # --- Step 4: Update Centroids ---
    # Move centroid to the mean of its assigned points
    new_centroids = np.zeros_like(centroids)
    for ki in range(k):
        # Only update if the cluster has points
        if len(data[groupidx == ki]) > 0:
            new_centroids[ki, :] = np.mean(data[groupidx == ki], axis=0)
        else:
            new_centroids[ki, :] = centroids[ki, :] # Keep old position if empty

    centroids = new_centroids

    # --- Visualization for this Step ---
    plot_kmeans_step(axes[i+1], data, centroids, groupidx, f"Iteration {i+1}")

plt.tight_layout()
plt.show()
```
